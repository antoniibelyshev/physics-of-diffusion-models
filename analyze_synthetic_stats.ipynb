```python
from stats import compute_stats
import torch
import matplotlib.pyplot as plt

D = 10
scales = 1 + torch.randn(D) * 0.1

temp = torch.logspace(-4, 4, 100)
hc_lst = []
n_lst = 10 ** torch.arange(2, 7)
for n in n_lst:
    y = torch.randn(n, D)
    y /= y.square().sum(1, keepdims=True).sqrt()
    y *= scales
    stats = compute_stats(y, temp, batch_size=100, n_iter=10)
    hc_lst.append(stats["var_H"] / temp.pow(2))

    plt.figure()
    plt.plot(temp, hc_lst[-1])
    plt.xscale("log")
    plt.show()
```
